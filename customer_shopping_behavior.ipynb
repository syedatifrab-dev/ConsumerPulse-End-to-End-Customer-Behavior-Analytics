```python
import pandas as pd
```

```python
df=pd.read_csv('customer_shopping_behavior.csv')
```

```python
df.head(10)
```

```python
df.isnull().sum()
```

```python
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns=df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
```

```python
df.columns
```

```python
df.columns = (
    df.columns
      .str.replace('_', '', regex=False)
      .str.lower()
      .str.replace(' ', '_')
)
```

```python
df.columns
```

```python
df = df.rename(columns={
    'customerid': 'customer_id',
    'itempurchased': 'item_purchased',
    'purchaseamount_usd': 'purchase_amount_usd',
    'reviewrating': 'review_rating',
    'subscriptionstatus': 'subscription_status',
    'shippingtype': 'shipping_type',
    'discountapplied': 'discount_applied',
    'promocodeused': 'promo_code_used',
    'previouspurchases': 'previous_purchases',
    'paymentmethod': 'payment_method',
    'frequencyofpurchases': 'frequency_of_purchases'
})
```

```python
df.columns
```

```python
df = df.rename(columns={'purchaseamount(usd)': 'purchase_amount'})
```

```python
df.columns
```

#create a cloumn age group


```python
labels= ['young Adult', 'Adult', 'Middle-Age', 'Senior']
df['age_group']=pd.qcut(df['age'],q=4, labels=labels)
```

```python
df[['age','age_group']].head(10)
```

```python
#create columns purchase_frequency_days

# create new column purchase_frequency_days

frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
df[['discount_applied','promo_code_used']].head(10)
```

```python
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
# Dropping promo code used column

df = df.drop('promo_code_used', axis=1)
```
